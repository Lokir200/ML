Min-Max Normalization

```python
import numpy as np
import pandas as pd
import math 
X=[1,3,5,6,7,8,21,31,31,37,12,509]
Y=[]
min=1
max=509
for i in range(len(X)):
    Y.append((X[i]-min)/(max-min))
print(Y)    
    
```

Decimal Scaling

```python
X=[1,3,5,6,7,8,21,31,31,37,12,509]
d=2
Y=[]
for i in range(len(X)):
    Y.append(X[i]/10**d)
print(Y)    
    
```

```python
X=[1,3,5,6,7,8,21,31,31,37,12,509]
Y=[]
sstd=0
mean=sum(X)/len(X)
for i in range(len(X)):
    sstd=sstd+((X[i]-mean)**2) 
var=sstd/len(X)
print("var",var)
std=np.sqrt(var)
print("std",std)
```

```python
Y=[]
for i in range(len(X)):
   Y.append((X[i]-mean)/std)
print(Y)


```

```python
import numpy as np
A=[1,3,np.nan,3,5,6,7,8,21,31,31,37,12,509]
print(A)
```

NAN mean

```python

sum=0
count=0
rem=0
for i in range(len(A)):
    if np.isnan(A[i])==True:
        rem=i
        print("none value sorted in index",i)
        count+=1
    else:
        sum+=A[i]
mean=sum/(len(A)-count)
A[rem]=mean
print(A)
```
